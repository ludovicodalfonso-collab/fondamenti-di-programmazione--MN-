## Set (Insiemi)

Un set è una struttura dati come una lista, ma che non ammette duplicati


````python
a = {1,2,3}
a.add(2)
print(a)

# trova elementi senza duplicati
x = [1,1,1,1,2,3,1,2,3,4]
a = set(x)
print(a)

# trova gli elementi che hanno duplicati
dup = set([y for y in x if x.count(y) > 1])
print(dup)

#trova elementi che non hanno duplicati
ndup = a.difference(dup) #intersezione, unione, ...
print(ndup)
````

### Serializzazione

* La **serializzazione** è un processo per salvare una variabile su un file (o per trasmetterlo su una connessione di rete).
* La **de-serializzazione** è il processo opposto. Legge un file (o dati provenienti dalla rete) e ricrea la la variabile

* In generale potremmo inventare noi un formato con cui slavare delle variabili su un file (serializzazione), poi leggere quel file e ricreare le variabili, ma ci sono librerie che lo fanno già

* JSON (JavaScript Object Notation) è uno standard per la memorizzazione delle informazioni (molto usato per le applicazioni Web)
* La libreria/modulo json offre una serie di funzionalità per serializzare/deserializzare variabili in json

````python
import json
a = {"nome":"marco", "anni":20}
with(open('data/a.json','w')) as ofile:
    json.dump(a,ofile)

````

````python
# cerca i file che contengono una parola in una cartella e sue sottocartelle
# usa una funzione ricorsiva

import os
def cerca_file(cartella, parola):
    for elemento in os.listdir(cartella):
        percorso = os.path.join(cartella, elemento) # f'{cartella}/{elemento}'
        if os.path.isfile(percorso):
            with open(percorso, 'r', encoding='utf-8', errors='ignore') as file:
                contenuto = file.read()
                if parola in contenuto:
                    print(f"Trovato in: {percorso}")
        elif os.path.isdir(percorso):
            cerca_file(percorso, parola)
# Esempio di utilizzo
cerca_file('../data/dir1', 'marco')
````

### Esercizio 1

Dato una lista di utenti con **username**, **email** ed **età**
Voglio creare una lista di controlli (realizzati tramite funzioni):
- username non vuoto 
- email deve contenere il carattere @
- età > 0 

Una funzione applica la lista di controlli e raccoglie tutti gli eventuali errori

````python
def username_non_vuoto(username, email, eta):
    if not username:
        print("Il nome utente non può essere vuoto.")
        return False
    return True
def at_nel_email(username, email, eta):
    if "@" not in email:
        print("Email non valida, manca '@'.")
        return False
    return True
def eta_valida(username, email, eta):
    if eta < 0 or eta > 120:
        print("Età non valida.")
        return False
    return True

controlli = [username_non_vuoto, at_nel_email, eta_valida]

def valida_utente(username, email, eta):
    for controllo in controlli:
        if not controllo(username, email, eta):
            return False
    return True

valida_utente("", "marcoe@xample.com", 25)
````

### Lambda

Le lambda sono una tecnica per creare delle funzioni senza dover usare la sintassi def

````python
#def add(x, y):
#    return x + y

add = lambda x,y: x+y
print(add(3,4))
print((lambda x,y: x+y)(3,4))

# Ternary Operators
pos_neg = lambda x: "positivo" if x > 0 else "negativo"
print(pos_neg(-1))

def multiplier(k):
    return lambda x:k*x

double = multiplier(2)
triple = multiplier(3)
print(double(4))
print(triple(3))

# Non tutte le istruzioni possono andare dentro una lambda. 
# Ad esempio non posso fare assegnamenti
````

###  Map, Reduce, Filter
* Sono funzioni per applicare funzioni a una lista 
* Utilizzo simile alle list comprehension

````python
def f(x):
    return x+1

x = [f(i) for i in [1,2,3]]
x = list(map(f,[1,2,3]))
print(x)


def f(x,y,z=0):
    print('sommo',x,y,z)
    return x+y+z

import functools
x = functools.reduce(f,[1,2,3])
print(x)


def f(x):
    return x >= 0

x = list(filter(f,[-1,0,1]))
print(x)
````

````python
def wrap(f):
    def newf(x):
        with open('log.txt','a') as logfile:
            logfile.write('Chiamata a funzione {}\n'.format(f.__name__))
        f(x)
    return newf

def stampa(x):
    print(x)

stampa = wrap(stampa)

# -------
stampa('ciao')
for i in range(3):
    stampa('ciao')

````

````python
def wrap(f):
    def newf(*args):
        print("Oprazione da fare prima")
        # es., authorization, logging
        f(*args)
    return newf

@wrap
def add(x,y):
    print(x+y)

@wrap
def stampa(x):
    print(x)


add(3,4)
stampa('Marco')
````

````python
def login_required(f):
    def decf():
        pw = input('inserisci password ')
        if pw == 'marco':
            f()
    return decf

@login_required
def simple_print():
    print('simeple print')

if __name__ == '__main__':
    simple_print()
````

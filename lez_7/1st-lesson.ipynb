```python
# SLIDE NO: 28
# This is a Guess the Number game.
import random

guessesTaken = 0

print('Hello! What is your name?')
myName = input()

number = random.randint(1, 20)
print('Well, ' + myName + ', I am thinking of a number between 1 and 20.')

for guessesTaken in range(6):
    print('Take a guess.') # Four spaces in front of "print"
    guess = input()
    guess = int(guess)
    
    
    if guess < number:
        print('Your guess is too low.') # Eight spaces in front of "print"

    if guess > number:
        print('Your guess is too high.')
    
    if guess == number:
        break
    
if guess == number:
    guessesTaken = str(guessesTaken)
    print('Good job, ' + myName + '! You guessed my number in ' +
          guessesTaken + ' guesses!')

if guess != number:
    number = str(number)
    print('Nope. The number I was thinking of was ' + number + '.')
```

```python
# SLIDE NO: 68
# exercise 1
# Write a script that accepts an input consisting of an integer representing a radius
# calculates the area of the corresponding circle, and prints the result
from math import pi
print("Input a number representing the radius of a circle, I'll calculate the Area")

while True:
    try:
        radius = input()
        radius = float(radius)
        break
    except ValueError:
        print("Please enter an integer or a float")

area = pi * radius ** 2
print("The area of a circle with radius " + str(radius) + " is: " + str(area))
```

```python
# SLIDE NO: 68
# exercise 2
# Write a script that accepts three integers. 
# If they are all different, print the sum of the three. 
# If two are the same, print the result of dividing the sum of the equal ones 
# by the third. 
# If all three are the same, print the result of (n + n)n.

print("Input three numbers I'll do some magic")

while True: # instead of doing three of these you can maybe use a list and a for number in range(3)
    try:
        print("Enter the first integer")
        first = input()
        first = int(first)
        break
    except ValueError:
        print("Please enter an integer")
while True:
    try:
        print("Enter the second integer")
        second = input()
        second = int(second)
        break
    except ValueError:
        print("Please enter an integer")
while True:
    try:
        print("Enter the third integer")
        third = input()
        third = int(third)
        break
    except ValueError:
        print("Please enter an integer")

if first != second and first != third and second != third:
    result = first + second + third
    print("The sum of the three of them is: " + str(result))
elif first == second == third:
    n = first
    result = (n + n) * n
    print("Printing the result of (n+n)n: " + str(result))
else:
    if first == second:
        equal_number = first
        different_number = third
    elif first == third:
        equal_number = first
        different_number = second
    elif second == third: # could have been simply else
        equal_number = second
        different_number = first
    result = (equal_number * 2) / different_number
    print("Printing the sum of equals divided by the different one: " + str(result))
```

```python
# SLIDE NO: 68
# exercise 3
# Write a script that accepts two integers. 
# If the value of their sum, or of their difference is 5, print "True";
# otherwise, print "False." 
# If the inputs entered are not numeric, print "Input must be a number."

print("Input two integer")
while True: # instead of doing two of these you can maybe use a list and a for number in range(2)
    try:
        print("Enter the first integer")
        first = input()
        first = int(first)
        break
    except ValueError:
        print("Please enter an integer")
while True:
    try:
        print("Enter the second integer")
        second = input()
        second = int(second)
        break
    except ValueError:
        print("Please enter an integer")
input_sum = first + second
input_diff = abs(first - second) # to avoid negative numbers lets use abs() --> absolute
if input_sum == 5 or input_diff == 5:
    print(str(True))
else:
    print(str(False))
```

```python
# SLIDE NO: 69
# exercise 4
# Write a script that accept input for name and surname,
# and print the string with characters at odd indices in reverse order
from string import ascii_letters # YES I know! we are not considering accent mark or symbols outside ascii

def check_if_ascii_letters_only(input_string):
    for letter in input_string:
        if letter not in ascii_letters:
            raise ValueError("Entered value has literarls that are not ascii letters")

print("Input a name and a surname")
while True:
    try:
        print("Enter a name")
        name = input()
        check_if_ascii_letters_only(name)
        break
    except ValueError:
        print("Please enter a name consisting of only ascii characters")

while True:
    try:
        print("Enter a surname")
        surname = input()
        check_if_ascii_letters_only(surname)
        break
    except ValueError:
        print("Please enter a surname consisting of only ascii characters")

fullname = name + surname # lets avoid the space in the middle
# odd_letters = fullname[1::2]
# print(odd_letters[::-1])
print(fullname[-2::-2])
```

```python
# SLIDE NO: 69
# exercise 7
# Write a script that calculate the difference between a given number and 17. 
# If the number is less than 17, print double the absolute difference.
# (Python has a built-in function called abs(n) that return the absolute value of an int or a float)

while True:
    try:
        print("Enter an integer")
        n = input()
        n = int(n)
        break
    except ValueError:
        print("Please enter an integer")
result =  abs(n - 17)
if n < 17:
    result *= 2
    
print(str(result))
```

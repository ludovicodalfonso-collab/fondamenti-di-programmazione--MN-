    <p>with built-in type python behave in an unexpected way, for example when we say a=1, b=1 and then ask if a is b</p>
    <p>teoretically no but practically the <strong>is</strong> keyword is used to test if two variables <strong>refer</strong> to the same object.</p>
</div>

```python
a = 1
b = a
assert a is b
```

```python
a = 1
b = 1
assert a is b #refer to same 1
```

```python
a = 'string'
b = a
assert a is b #refer to same 'string'
```

```python
a = 'string'
b = 'string'
assert a is b # refers to same 'string'
```

```python
a = [1]
b = [1]
assert a is b #refer to 2 different list whose first element refer to the same object
```

```python
# if we define our class 
class StrangeClass(object):
    def __init__(self, value):
        self.value = value
#     def __repr__(self):
#         return f'{self.__class__}: {self.value}'
    def __str__(self):
        return f'{self.value}'
a = StrangeClass(1)
b = StrangeClass(1)
assert a is b # refer to 2 object whose attribute value refer to the same object
```

    <p>strings are <strong>IMMUTABLE</strong> object. You may have the illusion of mutability due to add operation but the truth is that the result of an addition of string is a new string object</p>

```python
# pretty shabby example  to demostrate that string are IMMUTABLE
from string import ascii_uppercase
x = 'francesco'
if x[-1] not in ["!", "?", "."]:
    x = x + '!'
print(x)
if x[0] not in ascii_uppercase:
    x[0] = x[0].upper()
#     x = x[0].upper() + x[1:]
print(x)
```

## GENERATORS
<div class="alert alert-info" role="alert" style="margin:10px;">

    <p>Playing with generators, understanding their real use
</p>
</div>

```python
#generatori
def gen():
    lista = ['test_a', 'test_b', 'test_c', 'test_d']
    i = 0
    at_most = 100
    while True:
        yield lista[i%len(lista)]
        i +=1
        if i >= at_most:
            raise StopIteration
    
```

```python
iter_gen = iter(gen())
iter_gen
```

```python
try:
    next(iter_gen)
except StopIteration:
    print('done!')
```

```python
for el in gen():
    print(el)
```

```python
# duck typing mode on: un dizionario ha bisogno di __hash__ e __eq__ per utilizzare un oggetto come chiave
# DUCK TYPING da Alex Martelli: Don't check whether it is-a duck: check whether it quacks-like-a duck, 
# walks-like-a duck, etc, etc, depending on exactly what subset of duck-like behavior you need to play your
# language-games with.

class SuperInt:
    def __init__(self, integer, stringa=None):
        if type(integer) != int:
            raise TypeError
        self.value = integer
        self.stringa = stringa or str(integer)
    
    def __str__(self):
        return f'{self.stringa}'
    
    def __hash__(self):
        return self.value.__hash__()
    
    def __eq__(self, other):
        if type(other) == int:
            return self.value == other
        if type(other) == SuperInt:
            return self.value == other.value
        return False
    
    def __iadd__(self, other):
        if type(other) == int:
            self.value += other
            self.stringa = str(self.value)
            return self.value
        if type(other) == SuperInt:
            self.value += other.value
            self.stringa = str(self.value)
            return self.value
        raise TypeError
    
    def __add__(self, other):
        if type(other) == int:
            return SuperInt(self.value + other)
        if type(other) == SuperInt:
            return SuperInt(self.value + other.value)
        raise TypeError
```

```python
# example inspired by the Ramalho's example from Fluent Python
# concrete example of Duck-Typing
import random
class BingoBag:
    def __init__(self, numbers= range(1,91)):
        self._items = list(numbers) # _items is private -> its not a security concern,
                                    # it's only a warning for other developer using your code
                                    # it's only saying "wait before you use it lightly know that it's private
                                    # hence it may be used internally in the class and you modifying it may cause unwanted behaviour
        random.shuffle(self._items)
    
    def pick(self):
        try:
            return self._items.pop() # "pops" out the first item
        except IndexError: # done
            raise LookupError('The bag is empty!')
    
    def __call__(self): # __call__ is the magic method that make a class callable (normally only functions are collable)
        return self.pick() # in the end I'm simply creating a shortcut for self.pick()
    
    def __len__(self):
        return len(self._items)
    
    # __getitem__ makes BingoBag iterable
    def __getitem__(self, position): # it seems nosense for a BingoBag I know! (It would make much more sense on a card deck!)
        return self._items[position] 
    
```

```python
bag = BingoBag(range(1,91))
```

```python
bag.pick()
```

```python
bag() # chiamo sacco come fosse una funzione
```

```python
# difference between class and self
# static and class method how to -> https://dzone.com/articles/definitive-guide-how-use
class Demo:
    def nmethod(*args): # def nmethod(self, *args)
        return args

    @classmethod
    def cmethod(*args): # def cmethod(Demo, *args)
        return args

    @staticmethod
    def smethod(*args): # def smethod(*args)
        return args
    
def omethod(*args):
    return args
```

```python
Demo.nmethod('only this string: class is not yet instanciated hence self does not exist yet')
```
